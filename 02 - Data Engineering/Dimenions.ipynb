```python

## Dim_Educacion
# Leer la tabla bronze
df_bronce = spark.table("testetlfincas.bronze.bronce_fincas")

# Obtener valores únicos de EDUCACION y asignar id incremental
from pyspark.sql.functions import monotonically_increasing_id

df_dim_educacion = df_bronce.select("EDUCACION").distinct() \
    .withColumn("id_Educacion", monotonically_increasing_id())

# Crear la tabla Dim_Educacion
df_dim_educacion.write.format("delta").mode("overwrite").saveAsTable("testetlfincas.silver.Dim_Educacion")

display(df_dim_educacion)
```

```python
## Dim_Genero

df_dim_genero= df_bronce.select("GENERO").distinct() \
    .withColumn("id_Genero", monotonically_increasing_id())

# Crear la tabla Dim_Genero
df_dim_genero.write.format("delta").mode("overwrite").saveAsTable("testetlfincas.silver.Dim_Genero")

display(df_dim_genero)
```
